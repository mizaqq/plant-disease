```python
import os
from pathlib import Path

from src.utils.features import get_sample_images, show_annotations
```

```python
data_path = Path('.').joinpath('data')
annotation_path = data_path.joinpath('annotation/YOLO')
leaves_txt_list = [annotation_path.joinpath('leaves').joinpath(f) for f in os.listdir(annotation_path.joinpath('leaves'))]
leaves_image_list = [str(f) for f in data_path.joinpath('leaves').rglob('*.jpg') if f.is_file()]
```

```python
leaves_image_text = get_sample_images(leaves_txt_list, leaves_image_list)
```

```python
show_annotations(leaves_image_text)
```

```python
fruits_txt_list = [annotation_path.joinpath('fruits').joinpath(f) for f in os.listdir(annotation_path.joinpath('fruits'))]
fruits_image_list = [str(f) for f in data_path.joinpath('fruits').rglob('*.jpg') if f.is_file()]
```

```python
fruits_image_text = get_sample_images(fruits_txt_list, fruits_image_list)
show_annotations(fruits_image_text)
```
