```python
import os
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
data_path = Path('.').joinpath("data").joinpath("leaves")
classes = os.listdir(data_path)
```

```python
class_count=[]
for c in classes:
    class_path = data_path.joinpath(c)
    counta = len(os.listdir(class_path))
    class_count.append(counta)
```

```python
plt.figure(figsize=(8, 5))
plt.bar(classes, class_count, color=['blue', 'green', 'orange', 'red'], alpha=0.8, edgecolor='black')
plt.title('Category Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Categories', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.tight_layout()
plt.show()
```
